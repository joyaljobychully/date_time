?1) Print the current date

```python
from datetime import date
```

```python
print(date.today())
```

?2) Print the present time.

```python
from datetime import datetime
print(datetime.now())
```

?3) Subtract one day from today's date

```python
from datetime import timedelta
```

```python
print(date.today() - timedelta(days=1))
```

?4) Add one hour to the present time

```python
print(datetime.now() + timedelta(hours=1))
```

?5) Print the present IST.

```python
print(datetime.now() + timedelta(hours=5) + timedelta(minutes = 30))
```

?6) Add 35 seconds to present time.

```python
print(datetime.now() + timedelta(seconds = 35))
```
